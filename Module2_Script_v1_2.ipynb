## Clean up the data
Did you notice anything unusual about the "house" amounts? 

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Slice the data  
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2) 
```

## Some random stuff 
 Average income at the two levels?

 Conclusion - Higher income leaves more often than they stay.

```python
df_clean[['income','leave']].groupby('leave').mean().round(2)
mpl.bar(df['id'], df['income'])
mpl.xlabel('Leave Type')
mpl.ylabel('Average Income')
mpl.title('Average Income by Leave Type')
```

## More random cool stuff 
 Correlation amount of minutes on call with income?

 Conclusion - no correlation. amount of minutes between leave/stay individuals are very similar and thus don't correlate with income

```python
df_clean[['average_call_duration', 'leave']].groupby('leave').mean().round(2)

mpl.bar(df['id'], df['average_call_duration'])
mpl.xlabel('Leave Type')
mpl.ylabel('Average Call Duration')
mpl.title('Average Call Duration by Leave Type')
```
